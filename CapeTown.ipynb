#### Viewing population in Cape Town Suburbs

```python
import pandas as pd
import folium
import requests
import urllib.parse
from geopy.geocoders import Nominatim

def GetLatAndLongFromAddress(address):
	geolocator = Nominatim(user_agent="Your_Name")
	try:
	 location = geolocator.geocode(address)
	 if location.address.find('Cape Town') != -1:
		 return (location.latitude, location.longitude)
	 else:	 
		 return "" 
	except:
	 return ""

	
cptpopdata = pd.read_csv("capetownpopulationdata.csv")
cptpopdata
```

```python
latandlongdata = []
for i,row in cptpopdata.iterrows():
	data = GetLatAndLongFromAddress(row['name']+",South Africa")
	try:
		if len(str(data)) > 1:
			latandlongdata.append(data)
		else:
			cptpopdata.drop(i, inplace=True)
	except:
		pass
cptpopdata['latandlong'] = latandlongdata
cptpopdata
```

```python
import folium
import numpy as np
from folium import plugins
from folium.plugins import HeatMap
m = folium.Map(location=[-33.9249, 18.4241], tiles="OpenStreetMap", zoom_start=12)
```

```python
lat = [] 
long = []
pop = []
cptpopdata['population'] = cptpopdata['population'].astype(float)

for i,row in cptpopdata.iterrows():
	folium.Marker(location=[tuple(row['latandlong'])[0], tuple(row['latandlong'])[1]],popup=row['name']+"\n"+"Population is {0}".format(str(row['population'])),).add_to(m)
	lat.append(tuple(row['latandlong'])[0])
	long.append(tuple(row['latandlong'])[1])
	pop.append(row['population'])
   
            
cptpopdata['lat'] = lat
cptpopdata['long'] = long



data = cptpopdata[['lat', 'long','population']].values
m.add_child(plugins.HeatMap(data, radius=100))
 
m
```
